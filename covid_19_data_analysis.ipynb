```python
# Data Cleaning
def clean_data(df, state='California'):
    df = df[df['Province_State'] == state].set_index('Admin2').T
    drop_columns = ['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key']
    if 'Population' in df.index:
        drop_columns.append('Population')
    df.drop(drop_columns, errors='ignore', inplace=True)
    return df
```

```python
cases_clean = clean_data(cases)
cases_clean.head()
```

```python
deaths_clean = clean_data(deaths)
deaths_clean.head()
```

```python
# Process Population Data
pop["CTYNAME"] = pop["CTYNAME"].str.replace(" County", "")
pop.set_index('CTYNAME', inplace=True)
```

```python
pop.head()
```

```python
# Normalize Data Per Million
def normalize_per_million(df, pop_data):
    for county in df.columns:
        if county in pop_data.index:
            df[county] = (df[county] / pop_data.loc[county, 'Pop']) * 1_000_000
        else:
            print(f"Population data missing for {county}")
    return df
```

```python
cases_pm = normalize_per_million(cases_clean.copy(), pop)
deaths_pm = normalize_per_million(deaths_clean.copy(), pop)
```

```python
# Fixing Data Anomalies
def fix_anomalies(df):
    # Replacing negative values
    df[df < 0] = None
    # Smoothing sudden spikes (3-day rolling median) and ensuring dtype is float
    return df.ffill().astype(float).rolling(window=3, min_periods=1).median()
```

```python
# Applying Fix to Cases and Deaths Data
cases_clean = fix_anomalies(cases_clean)
deaths_clean = fix_anomalies(deaths_clean)
```

```python
cases_7d = fix_anomalies(cases_clean.diff().rolling(window=7).mean())
deaths_7d = fix_anomalies(deaths_clean.diff().rolling(window=7).mean())
cases_pm_7d = fix_anomalies(cases_pm.diff().rolling(window=7).mean())
deaths_pm_7d = fix_anomalies(deaths_pm.diff().rolling(window=7).mean())
```

```python
# Plot Function
def plot_data(df, title, counties=None, top_n=None):
    if top_n:
        latest_date = df.index[-1]
        df = df.sort_values(by=latest_date, axis=1, ascending=False).iloc[:, :top_n]
    elif counties:
        df = df[counties]

    df.plot(figsize=(12, 6), title=title)
    plt.show()
```

```python
selected_counties = ['Alameda', 'San Francisco', 'San Mateo', 'Santa Clara']
```

# Visualization

```python
plot_data(cases_clean, "COVID-19 Cases in Selected California Counties", selected_counties)
```

```python
plot_data(deaths_clean, "COVID-19 Deaths in Selected California Counties", selected_counties)
```

```python
plot_data(cases_7d, "7-Day Average New COVID-19 Cases in Selected Counties", selected_counties)
```

```python
plot_data(deaths_7d, "7-Day Average New COVID-19 Deaths in Selected Counties", selected_counties)
```

```python
plot_data(cases_pm_7d, "Top 10 Counties by 7-Day Avg New Cases Per Million", top_n=10)
```

```python
plot_data(deaths_pm_7d, "Top 10 Counties by 7-Day Avg New Deaths Per Million", top_n=10)
```
